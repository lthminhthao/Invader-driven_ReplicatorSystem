```python
import mpmath as mp
```

```python
def compute_analytical_terms():
    results = {}

    # N = 3
    results['N=3, k=2'] = 2 - 2 * mp.log(2)
    results['N=3, k=3'] = -1 + 2 * mp.log(2)

    # N = 4
    results['N=4, k=2'] = 9/2 - 6 * mp.log(2)
    results['N=4, k=3'] = (-2 * mp.pi**2 + 42 * mp.log(2) - 12  - 24 * mp.log(2)**2 - 18 * mp.log(3)  - 24 * mp.polylog(2, -2))
    
    results['N=4, k=4'] = (2 * mp.pi**2 - 36 * mp.log(2) + 17/2  + 24 * mp.log(2)**2 + 18 * mp.log(3) + 24 * mp.polylog(2, -2))

    # N = 5 
    results['N=5, k=2'] = 17/2 - 12 * mp.log(2)
    results['N=5, k=3'] = (-105 - 24 * mp.pi**2 + 420 * mp.log(2) - 288 * mp.log(2)**2 - 204 * mp.log(3) - 288 * mp.polylog(2, -2))
    
    return results
```

```python
if __name__ == "__main__":
    computed_terms = compute_analytical_terms()
    
    print("Analytical probabilities (%):")
    print("-" * 30)
    for key, value in computed_terms.items():
        prob_real = float(value.real)
        percentage = prob_real * 100
        
        print(f"{key}: {percentage:.2f}%")
```
